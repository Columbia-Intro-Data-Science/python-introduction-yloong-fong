e) Evalute the R^2 on **testing** data. Is this good? Bad? Why?

````python
print('R^2 Score: %.2f' % regr.score(X_test, y_test))
````

##### Comments: 

The $R^2$ formula is:

$
\begin{equation}
R^2 = 1 - \frac{\sum_{i=1}^N (y_i - \hat y_i)^2}{\sum_i (y_i - \bar y)^2}
\end{equation}
$

where $\hat y_i$ is our predicted sales, and $y_i$ is the actual sales. The $R^2$ therefore reveals the extent to which the variance in actual sales of the testing data can be explained by the model that we derived from the training data. The $R^2$ is scaled such that the maximum possible $R^2$ (which implies all variance is explained) is 1. Here we have an $R^2$ of 0.88 which is means that our model explains 92% of the variance. That suggests that our model is quite good. Because the $R^2$ is not too high, it is also unlikely that we have model has been over-fitted.

## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations, learning_rate = 0.0001):
    points = genfromtxt("../Documents/GitHub/APMAE4990-/data/data.csv", delimiter=",")
    learning_rate = learning_rate
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print("Running...")
    
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1],color='b')
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

run(100)
````

##### Comments: 

Gradient descent allows us to find a function of our X features that minimises our $\mathcal{L}_p$ norms, such that the predicted output values most closely fit our actual output values. The learning rate determines the size of the "steps" that the gradient descent method uses in each iteration that attempts to minimise the $\mathcal{L}_p$ norms.

Hence, if the learning rate is small, a larger number of small "steps" will be needed to find the function that minimises our $\mathcal{L}_p$ norms. This will necessarily mean that convergence to the actual minimum point will take longer. By contrast, if the learning rate is too large, the "steps" taken might be too big and the method may bypass the actual minimum point altogether. In this circumstance, over-estimation continually recurs and hence the solution from the method will diverge from the actual minimum point. 

We can tell how well the solution from our method converges with the function that minimises our $\mathcal{L}_p$ norms by looking at the error. In the case of the above, with a learning rate of 0.0001 and after 100 iterations, the given solution only has an error of 112.647. This is very good because the error is much smaller than the initial error of 5565.108, when the model assumed a function of X features that has a gradient and intercept of 0.

````python
for num in range(0,10):
    run(num, 0.0001)
    plt.show()
````

````python
for num in range(0,10):
    run(num, 0.00001)
    plt.show()
````

````python
for num in range(0,10):
    run(num, 0.001)
    plt.show()
````

##### Comments: 

From the simulations above, we can see that different learning rates lead to different degrees of convergence after 10 iteration. The first learning rate I test is 0.0001. This seems to be the best learning rate because the error in the solution after 10 iterations is only 112.672. 

The second learning rate I test is 0.00001. This is a smaller learning rate that the previous one, and so we expect the gradient descent method to be more inefficient and for convergence to take longer. Indeed, the error in this solution after 10 iterations is higher at 2285.085. However, we know convergence is still occuring as the error is still smaller than the initial error of 5565.108.

The final learning rate I test is 0.001. This is a larger learning rate that the first learning one, and so we expect the gradient descent method might continually overestimate the minimum point and produce divergent solution.Indeed, the error in this solution after 10 iterations is very much higher at 345950151.478.

## Problem 3 Part 2

Plot the error as a function of the number of iterations for various learning rates. Choose the rates
so that it tells a story.

````python
#defining components for gradient_descent_runner function
points = genfromtxt("../Documents/GitHub/APMAE4990-/data/data.csv", delimiter=",")
starting_b = 0 
starting_m = 0 
num_iterations = 3

#defining learning_rates to be plotted
learning_rates = np.arange(0.00005,0.0004,0.00001)

#creating new array for errors
errors=[]

#creatng forloop for learning_rates to generate b, m and errors for each learning rate
for x in learning_rates:
    learning_rate = x
    [b, m] = gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations)
    errors.append(compute_error_for_line_given_points(b, m, points))

#Plot error as function of 10 iterations for various learning rates
plt.xlabel('Learning Rate')
plt.ylabel('Error ')
plt.plot(learning_rates,errors)
````

##### Comments: 

From the graph, we can compare how the errors look like in solutions for various learning rates after 3 iterations. Based on the simulations earlier which found that the optimum learning rate is close to 0.0001, I choose a range of learning rates from 0.00005 (which will probably be inefficient) and 0.0004 (which will probably lead to overestimation)

The graph reveals a U-shaped curve. When the learning rate is less than 0.00015, the solutions produce errors higher than 0, revealing that convergence has yet to occur. When the learning rate is more than 0.00025, the solutions again produce errors higher than 0, and the errors produce increase as the learning rate increases, hence suggesting that such learning rates have divergent solutions.

From the graph, we further find that the optimum learning rate is not 0.0001 as we had initially guessed, but rather closer to 0.0002, which is the minimum point of the graph and therefore produces the least errors.
